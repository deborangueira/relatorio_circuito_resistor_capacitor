# Gráficos de análise do circuito Capacitor-Resistor

```python
# Importando as bibliotecas necessárias

import matplotlib.pyplot as plt
import pandas as pd
```

Para plotar esses gráficos, a base de dados utilizada é a gerada pelo circuito que está no arquivo `dados_cr.txt`. Nele há as informações de tempo, tensao do resistor e de tensao do capacitor retirados do Monitor Serial.

```python
df = pd.read_csv("dados_rc.txt", sep=r"\s+", names=["tempo", "Vc", "Vin"])
print(df)
```

### Carga no capacitor (Vc)
O gráfico mostra como a tensão no capacitor se comporta com o tempo (a curva de carga).

```python
plt.plot(df["tempo"], df["Vc"], color='blue', label='Tensão no Capacitor (Vc)')
plt.title("Carga no Capacitor")
plt.xlabel("Tempo (ms)")
plt.ylabel("Tensão (V)")
plt.grid(True)
plt.legend()
plt.show()
```
