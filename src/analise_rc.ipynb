# Análise Experimental do Circuito RC

Este notebook realiza a análise de um circuito RC (resistor–capacitor) a partir de dados coletados experimentalmente do monitor serial e salvos no arquivo `dados_rc.txt`. O conjunto de scripts lê as medições de tempo, tensão no capacitor (Vc), tensão de entrada (Vin) e calcula a tensão no resistor (Vr = Vin − Vc). Os resultados permitem visualizar o fenômeno de **filtragem** e **temporização** típico de circuitos RC, relacionando teoria e prática de eletrônica básica e programação embarcada.

Para fazer essa análise utilizarei python e bibliotecas para análise de dados e plotar imagens.

```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt
import pandas as pd

# Dataset
df = pd.read_csv("dados_rc.txt", sep=r"\s+", names=["tempo", "Vc", "Vin"])
print(df)
```

### Carga no capacitor (Vc)
O gráfico mostra como a tensão no capacitor se comporta com o tempo (a curva de carga).

```python
plt.figure()

plt.plot(df["tempo"], df["Vc"], color='blue', label='Tensão no Capacitor (Vc)')
plt.title("Carga no Capacitor")
plt.xlabel("Tempo (ms)")
plt.ylabel("Tensão (V)")
plt.grid(True)
plt.legend()
plt.show()

plt.savefig("grafico_carga.png", dpi=150, bbox_inches="tight")
plt.close()
```


### Descarga no resistor (Vr)

```python
df["Vr"] = df["Vin"] - df["Vc"] # Calcular a tensão no resistor

plt.figure()

plt.plot(df["tempo"], df["Vr"], color='red', label='Tensão no Resistor (Vr)')
plt.title("Descarga no Resistor (R)")
plt.xlabel("Tempo (ms)")
plt.ylabel("Tensão (V)")
plt.grid(True)
plt.legend()
plt.show()

plt.savefig("grafico_descarga.png", dpi=150, bbox_inches="tight")
plt.close()
```

### Comparativo (Vc x Vr)

```python
plt.figure()

plt.plot(df["tempo"], df["Vc"], color='blue', label='Tensão no Capacitor (Vc)')
plt.plot(df["tempo"], df["Vr"], color='red', label='Tensão no Resistor (Vr)')
plt.title("Comparativo: Carga e Descarga no Circuito RC")
plt.xlabel("Tempo (ms)")
plt.ylabel("Tensão (V)")
plt.grid(True)
plt.legend()
plt.show()

plt.savefig("grafico_comparativo.png", dpi=150, bbox_inches="tight")
plt.close()
```
